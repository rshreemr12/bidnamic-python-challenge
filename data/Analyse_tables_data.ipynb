```python
# Compute the ROAS and get top 10 searc
search_terms_df['ROAS'] = search_terms_df['conversion_value']/search_terms_df['cost']

search_terms_df.dropna().sort_values(by=['ROAS'], ascending=False).head(n = 10)

# return the Top 10 Search Terms by ROAS for a campaign structure_value 
# or adgroup alias.
```
